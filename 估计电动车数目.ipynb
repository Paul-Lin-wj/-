```python
import random
import matplotlib.pyplot as plt

# 生成1到500之间的30个随机整数
total_numbers = random.sample(range(1, 501), 30)
print("生成的30个随机整数为：", total_numbers)

# 初始化模型结果列表
model3_results1 = []
model4_results1 = []
model5_results1 = []

model3_results2 = []
model4_results2 = []
model5_results2 = []
```

```python
# 重复抽取10个数的步骤3次，并画出每次的样本分布图
for i in range(3):
    # 从这30个数中随机抽取10个数
    sampled_numbers = random.sample(total_numbers, 10)
    print(f"第{i+1}次从30个数中随机抽取的10个数为：", sampled_numbers)
    
    # 计算模型3的预测结果
    model3_result1 = sampled_numbers[-1] + sampled_numbers[0] - 1
    model3_results1.append(model3_result1)
    
    # 计算模型4的预测结果
    n = len(sampled_numbers)
    intervals = [j - i for i, j in zip(sampled_numbers[:-1], sampled_numbers[1:])]
    average_interval = sum(intervals) / n
    model4_result1 = (1 + 1/n) * (sampled_numbers[-1] - 1)
    model4_results1.append(model4_result1)
    
    # 计算模型5的预测结果
    model5_result1 = (1 + 1/(2*n-1)) * (sampled_numbers[-1] - 1/(2*n))
    model5_results1.append(model5_result1)
    '''
    # 画出样本分布图
    plt.figure(figsize=(10, 5))
    plt.hist(sampled_numbers, bins=range(1, 510, 50), alpha=0.7, label=f'Sample {i+1}')
    
    # 添加标题和图例
    plt.title('Sample Distribution')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    
    # 显示图形
    plt.show()
'''
# 计算三次样本的平均值作为模型估计的最终结果
final_model3_result1 = sum(model3_results1) / len(model3_results1)
final_model4_result1 = sum(model4_results1) / len(model4_results1)
final_model5_result1 = sum(model5_results1) / len(model5_results1)

print(f"模型3估计的最终结果：{final_model3_result1}")
print(f"模型4估计的最终结果：{final_model4_result1}")
print(f"模型5估计的最终结果：{final_model5_result1}")

# 画出三个模型的结果的对比图
plt.figure(figsize=(8, 6))
plt.bar(['Model 3', 'Model 4', 'Model 5'], [final_model3_result1, final_model4_result1, final_model5_result1], color=['blue', 'green', 'red'])
plt.title('Comparison of Model Results')
plt.xlabel('Model')
plt.ylabel('Estimated Total')
plt.show()
```
